Q1 :- How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain

For getting a specific (i,j) element of final matrix after multiplying two matrix:-

We have to do k times element wise multiplication of ith row of first matrix to jth column of second matrix and k-1 addition for total k elements.

total number of elements in final matrix = m*n

therefore, total number of multiplication = m*n*k
           total number of addition = m*n*(k-1)
           

Q2: -Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy.
Compare the timing of both solutions. Which one is faster? Why?

```python
#input
from numpy import random

n = int(input())
k = int(input())
m = int(input())

M1 = list()
M2 = list()

for i in range(n):
    test = list()
    for j in range(k):
        test.append(random.randint(1,100))
    M1.append(test)

for i in range(k):
    test = list()
    for j in range(m):
        test.append(random.randint(1,100))
    M2.append(test)
print(M1,M2)

M = list()
for i in range(n):
    test = list()
    for j in range(m):
        test.append(0)
    M.append(test)
```

```python
# Q2:- using list of list
import time

begin = time.time_ns()

for i in range(n):
    for j in range(m):
        for t in range(k):
            M[i][j] += M1[i][t]*M2[t][j]
            
end = time.time_ns()

time1 = (end-begin)/10**6

print("time taken for execution: ", time1 ,"ms")
print("\n Final Matrix: ", "\n", M)
```

```python
# Q2:- using numpy

import numpy as np

import time

begin = time.time_ns()

M = np.dot(M1,M2)

end = time.time_ns()

time2 = (end-begin)/10**6

print("time taken for execution: ", time2, "ms")
print(" \n Final Matrix: ", "\n", M)
```

```python
#Q2 : comparision of running time
print("time taken by program in case of list : ", time1)
print("time taken by program in case of numpy : ", time2)
```

Q10:- Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. 
Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as a, b, c, d, e. 
Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import pandas as pd
from numpy import random
z = np.random.uniform(low = 1, high = 2, size = (20,5))  
# print(z)

col = ['a','b','c','d','e']

df = pd.DataFrame(data = z, columns = col)
print(df,"\n")


std_dev = df.std()
maxm_std = max(std_dev)
print("column with highest standard deviation:", col[np.argmax(std_dev)], "\nvalue: ",maxm_std)



mean_row = df.mean(axis=1)
minm_mean = min(mean_row)
row_num = np.argmin(mean_row)
print("\nrow with minimum mean:", row_num,"\nvalue: ", minm_mean)

# print(mean_row)
# print(std_dev)
```

Q13:-Write a function to compute the argmin of a numpy array. The function should take a numpy array as 
input and return the index of the minimum element. You can use the np.argmin function to verify your solution.
